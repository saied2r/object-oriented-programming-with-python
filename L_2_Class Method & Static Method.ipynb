###Class Method & Static Method 
Python Methods


1.   Instance Method
2.   Class Method
3.   Static Method

**Class Method**

The class method takes a different parameter “cls” instead of “self” that points to the class instead of the object. Hence, it can’t modify the state of an instance but can modify the state of a class abd use decorator @classmethod.

 **Static Method**

The static method does not take any parameter like “cls” or “self”. it has nothing to do with the state of the class nor the state of the object. It uses decorator @staticmethod. If we want to do something extra with a class we can use static methods.




```python
class MyClass:
  def method(self):
    return 'instance method called', self
    
  @classmethod
  def classmethod(cls):
    return 'class method called', cls
  
  @staticmethod
  def staticmethod():
    return 'static method called'
```

```python
#Class method
from datetime import date
# random Person
class Person:
  total_age=0
  @classmethod
  def birthYear(cls, name, birthYear):
    print(Person.total_age)
    return name, date.today().year - birthYear

  def __init__(self, name, age):
    self.name = name
    self.age = age
    Person.total_age+=age
    
    
  def display(self):
    print(self.name + "'s age is: " + str(self.age))
    print(Person.birthYear(self.name,1997))
    
print(Person.birthYear('Padma',2005))
person = Person('Adam', 19)
person.display()
person1 = Person('Dama', 29)
print(Person.birthYear('Adma',1997))
person1.display()
print(Person.birthYear(person1.name,2000))
```

```python
#Static method
class Math:
  @staticmethod
  def Multiply(one, two):
    return one * two
    
math = Math()
if(12*72 == math.Multiply(12, 72)):
  print("Equal")
else: print("Not Equal")
```

```python
class Student:
  uni = "Bracu"    # Class variable
  def __init__(bleh, name, id):
    bleh.name = name
    bleh.id = id

  @classmethod
  def updateUni(cls,uni):   #Class method
    # Student.uni = uni
    cls.uni = uni

  def details(self):        # Instance method
    print("name:",self.name,"id:",self.id,"Uni name:",Student.uni)

  @staticmethod
  def checkDept(id):    # Static Method
    if id[3:5] == "01":
      print("CSE")
    elif id[3:5] == "41":
      print("CS")

s1 = Student("Bob", 11)
s2 = Student("Carol", 22)
s1.details()
s2.details()
Student.updateUni("Brac University")
s1.details()
s2.details()
Student.checkDept("14341007")
Student.checkDept("15301007")
s1.checkDept('17101379')
```

```python
#next class start
import math
class Cylinder:
  radius = 5
  height = 18

  def __init__(self, r, h):
    self.radius = r
    self.height = h
    print(f'Default radius={Cylinder.radius} and height= {Cylinder.height}.')
    Cylinder.radius = self.radius
    Cylinder.height = self.height
    print(f'Updated: radius = {self.radius} and height = {self.height}.')


  @staticmethod
  def area(r, h):
    print(f'Area: {2*math.pi*r**2+2*math.pi*r*h}')
  
  @staticmethod
  def volume(r, h):
    print(f'Volume: {math.pi*r**2*h}')

  @classmethod
  def swap(cls,r, h):
    return cls(h,r)

  @classmethod
  def changeFormat(cls,val):
    new = val.split('-')
    r, h = float(new[0]), float(new[1])
    return cls(r,h)

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13") # Cylinder(7,13)
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
class Student:
  uni = "Bracu"    # Class variable
  def __init__(bleh, name, id):
    bleh.name = name
    bleh.id = id

  @classmethod
  def updateUni(cls,uni):   #Class method
    # Student.uni = uni
    cls.uni = uni

  @classmethod
  def create(cls,var):
    n,i=var.split(',')
    return cls(n,int(i))

  def details(self):        # Instance method
    print("name:",self.name,"id:",self.id,"Uni name:",Student.uni)

  @staticmethod
  def checkDept(id):    # Static Method
    if id[3:5] == "01":
      print("CSE")
    elif id[3:5] == "41":
      print("CS")

s1 = Student("Bob", 11)
s2 = Student("Carol", 22)
s1.details()
s2.details()
Student.updateUni("Brac University")
s1.details()
s2.details()
Student.checkDept("14341007")
Student.checkDept("15301007")
s1.checkDept('17101379')
s3 = Student.create("Sheldon,1111") #Student("Sheldon",1111)
s3.details()
print(s3.name)
```
