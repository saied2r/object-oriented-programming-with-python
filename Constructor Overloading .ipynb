###Pass by Reference, pass by value, Method overloading


*   Whenever an immutable (int,string, float, tuple, etc.) value has been passed, it is called pass by value.You can assume that a copy of the value is passed here. 

*   On the other hand, whenever a mutable (List, dictionary) value or an object has been passed, rather than sending a copy of the values, we actually send the reference/location of the value.

###Method Overloading
When a class have multiple methods by same name but different parameters, i.e. known as method overloading. Python doesn't allow method overloading but we can achieve method overloading

```python
class Test:

  def __init__(self):
    print('Testing')
  
  def summation(self, a, b):
    print(a+b)
  
  def summation(self, a, b, c):
    print(a+b+c)

t1 = Test()
t1.summation(2,3)
t1.summation(1,2,3)
```

  
1.   Use default parameter in cases where you know you will get a certain number of arguments.



```python
class SumClass:
  def summation(self, a, b= None, c= None):
    if b!= None and c!= None:
       sum = a+b+c
    elif b!= None:
      sum = a+b
    else:
      sum = a
    
    print(sum)

s1 = SumClass()
s1.summation(2)    
s2 = SumClass()
s2.summation(2,4)
s3 = SumClass()
s3.summation(2,4,6)
```

2.   Use Asterisk sign $*$ to take all the arguments as a tuple and then iterate through a loop to get each element



```python
class SumClass:
  def sum(self,*n):
    sum = 0
    for i in n:
      sum = sum + i
    print(sum)
    
s1 = SumClass()
s1.sum(2,4)
s2 = SumClass()
s2.sum(2,4,6)
s2.sum(2,3,4,5,1,1)
```

3.  Use dispatch package (However, not suggested)

```python
#way 3. use dispatch package
from multipledispatch import dispatch
class SumClass:

  @dispatch(int,int)
  def sum(self,a,b):
    print(a+b,'int,int')

  @dispatch(int,int,int)
  def sum(self,a,b,c):
    print(a+b+c)

  @dispatch(str,str)
  def sum(self,a,b):
    print(int(a)+int(b))

  @dispatch(str,float)
  def sum(self,a,b):
    print(int(a)+b)

  @dispatch(int,str)
  def sum(self,a,b):
    print(a+int(b),'int,str')
  
s1 = SumClass()
s1.sum(2,4)
s1.sum(6,'77')
s1.sum(2,4,6)
s1.sum("2","3")
s1.sum("2", 3.0)
```
